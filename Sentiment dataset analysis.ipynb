```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset 
data = pd.read_csv(r"C:\Users\revan\Downloads\Sentiment dataset.csv")
```

```python
data.head()
```

```python
# Summary statistics for numerical columns
summary_stats = data.describe()
print("Summary Statistics:")
print(summary_stats)
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("\nMissing Values:")
print(missing_values)
```

```python
# Visualize the distribution of the 'Sentiment' column
plt.figure(figsize=(10, 6))
sns.countplot(x='Sentiment', data=data, palette='viridis')
plt.title('Distribution of Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize the distribution of 'Likes' using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(data['Likes'], bins=30, kde=True, color='blue')
plt.title('Distribution of Likes')
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize the distribution of 'Retweets' using a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['Retweets'], color='orange')
plt.title('Box Plot of Retweets')
plt.xlabel('Retweets')
plt.show()
```

```python
# Scatter plot to visualize the relationship between 'Likes' and 'Retweets'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Likes', y='Retweets', data=data, hue='Sentiment', palette='viridis')
plt.title('Scatter Plot: Likes vs Retweets')
plt.xlabel('Likes')
plt.ylabel('Retweets')
plt.show()
```

```python
# Correlation matrix for numerical features
numerical_features = data.select_dtypes(include=[np.number])
correlation_matrix = numerical_features.corr()
```

```python
# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Generate a report summarizing insights from the EDA
print("\nEDA Insights:")
print("1. The dataset contains a mix of positive, negative, and neutral sentiments.")
print("2. The 'Likes' and 'Retweets' columns show a positive correlation, indicating that posts with more likes tend to have more retweets.")
print("3. The distribution of 'Likes' is right-skewed, with most posts having a lower number of likes.")
print("4. The box plot of 'Retweets' shows some outliers, indicating a few posts with significantly higher retweets.")
print("5. The correlation matrix suggests that 'Likes' and 'Retweets' are moderately correlated.")
```
